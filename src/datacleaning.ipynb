```python
#import modules
import pandas as pd
```

```python
#convert the csv to dataframe for cleaning
df = pd.read_csv(r"C:\Users\DELL\Documents\dataluminaries\data\student_responses.csv")
df.head()
```

```python
#check for columns
df.info()
```

```python
df = df.drop('@', axis=1)
```

```python
df.info()
```

```python
df.info()
```

```python
mean_value = df['How often do you use the internet to access learning materials'].mean()
df['How often do you use the internet to access learning materials']=df['How often do you use the internet to access learning materials'].fillna(value=mean_value)
```

```python
df['How often do you use the internet to access learning materials']=df['How often do you use the internet to access learning materials'].astype(int)
```

```python
df.info()
```
